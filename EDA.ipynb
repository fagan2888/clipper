```python
df_fares = df.assign(Data_Value=df.FareAmount.abs())\
       .groupby(['PaymentProductName']).Data_Value.agg([('Min' , 'min'), ('Median', 'median'), ('Max', 'max'), ('Count', 'count')])\
       .add_prefix('PaymentProductName')
df_fares.sort_values('PaymentProductNameCount', ascending=False)
```

```python
df_fares = df.assign(Data_Value=df.FareAmount.abs())\
       .groupby(['PaymentProductName']).Data_Value.agg([('Min' , 'min'), ('Median', 'median'), ('Max', 'max'), ('Count', 'count')])\
       .add_prefix('PaymentProductName')
df_fares.sort_values('PaymentProductName').to_csv('data/fares.csv')
```
